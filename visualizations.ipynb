```python
import numpy as np
import matplotlib.pyplot as plt

# Create a matrix of scores between 6 batsmen
scores = np.array([[0, 71.98, 79.04, 67.06, 75.60, 82.50],
                   [71.98, 0, 73.14, 87.87, 80.13, 73.91],
                   [79.04, 73.14, 0, 70.65, 85.04, 91.30],
                   [67.06, 87.87, 70.65, 0, 84.52, 71.29],
                   [75.60, 80.13, 85.05, 84.52, 0, 84.17],
                   [82.50, 73.91, 91.30, 71.29, 84.17, 0]])

# Define the names of the batsmen
batsmen_names = ['Novel 1', 'Novel 2', 'Novel 3', 'Novel 4', 'Novel 5', 'Novel 6']

# Create a heatmap of the scores matrix
fig, ax = plt.subplots()
im = ax.imshow(scores)

# Set the ticks and tick labels for the axes
ax.set_xticks(np.arange(len(batsmen_names)))
ax.set_yticks(np.arange(len(batsmen_names)))
ax.set_xticklabels(batsmen_names)
ax.set_yticklabels(batsmen_names)

# Rotate the tick labels and set the axis labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
ax.set_xlabel('Novels')
ax.set_ylabel('Novels')

# Loop over the data and add the scores to the heatmap
for i in range(len(batsmen_names)):
    for j in range(len(batsmen_names)):
        text = ax.text(j, i, scores[i, j], ha="center", va="center", color="w")

# Add a title and show the plot
ax.set_title("Similarity Percentage between novels of Agatha Christie")
fig.tight_layout()
plt.show()
```

```python
import numpy as np
from tabulate import tabulate

# Create a matrix of scores between 6 batsmen
scores = np.array([[0, 71.98, 79.04, 67.06, 75.60, 82.50],
                   [71.98, 0, 73.14, 87.87, 80.13, 73.91],
                   [79.04, 73.14, 0, 70.65, 85.04, 91.30],
                   [67.06, 87.87, 70.65, 0, 84.52, 71.29],
                   [75.60, 80.13, 85.05, 84.52, 0, 84.17],
                   [82.50, 73.91, 91.30, 71.29, 84.17, 0]])

# Define the names of the batsmen
batsmen_names = ['Batsman 1', 'Batsman 2', 'Batsman 3', 'Batsman 4', 'Batsman 5', 'Batsman 6']

# Convert the matrix to a list of lists
scores_list = scores.tolist()

# Add the batsmen names as the first column and row of the table
for i in range(len(batsmen_names)):
    scores_list[i].insert(0, batsmen_names[i])
scores_list.insert(0, [''] + batsmen_names)

# Print the scores in a table format
print(tabulate(scores_list, headers='firstrow', tablefmt='orgtbl'))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a matrix of similarity percentages between 6 novels of Agatha Christie
scores = np.array([[0, 71.98, 79.04, 67.06, 75.60, 82.50],
                   [71.98, 0, 73.14, 87.87, 80.13, 73.91],
                   [79.04, 73.14, 0, 70.65, 85.04, 91.30],
                   [67.06, 87.87, 70.65, 0, 84.52, 71.29],
                   [75.60, 80.13, 85.05, 84.52, 0, 84.17],
                   [82.50, 73.91, 91.30, 71.29, 84.17, 0]])

# Define the names of the novels
novel_names = ['Novel 1', 'Novel 2', 'Novel 3', 'Novel 4', 'Novel 5', 'Novel 6']

# Create a heatmap of the similarity percentages matrix
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(scores, cmap='coolwarm')

# Add a colorbar legend to the plot
cbar = ax.figure.colorbar(im, ax=ax)

# Set the ticks and tick labels for the axes
ax.set_xticks(np.arange(len(novel_names)))
ax.set_yticks(np.arange(len(novel_names)))
ax.set_xticklabels(novel_names, fontsize=12)
ax.set_yticklabels(novel_names, fontsize=12)

# Rotate the tick labels and set the axis labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
ax.set_xlabel('Novels', fontsize=14)
ax.set_ylabel('Novels', fontsize=14)

# Loop over the data and add the similarity percentages to the heatmap
for i in range(len(novel_names)):
    for j in range(len(novel_names)):
        text = ax.text(j, i, scores[i, j], ha="center", va="center", color="w")

# Add a title to the plot and show it
ax.set_title("Similarity Percentages between Novels of Agatha Christie", fontsize=16)
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define the tabular data
data = [['', 'Col 1', 'Col 2', 'Col 3', 'Col 4'],
        ['Row 1', 1, 2, 3, 4],
        ['Row 2', 5, 6, 7, 8],
        ['Row 3', 9, 10, 11, 12],
        ['Row 4', 13, 14, 15, 16]]

# Convert the data to a Pandas DataFrame
df = pd.DataFrame(data[1:], columns=data[0])

# Create a figure and axis object
fig, ax = plt.subplots()

# Remove axis ticks and spines
ax.axis('off')

# Create the table
table = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')

# Set the font size of the table cells
table.set_fontsize(14)

# Set the background color of the table headers
for i in range(len(df.columns)):
    table[0, i].set_facecolor('lightblue')

# Set the background color of the table rows
for i in range(1, len(df)):
    for j in range(len(df.columns)):
        table[i, j].set_facecolor('whitesmoke')

# Add a title to the table
ax.set_title('Tabular Data Visualization', fontweight='bold', fontsize=18)

# Adjust the layout of the table
fig.tight_layout()

# Display the plot
plt.show()
```
